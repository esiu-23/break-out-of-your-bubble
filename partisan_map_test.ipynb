```python
# Plot only California
ca_source = GeoJSONDataSource(geojson = california.to_json())
# Source file for Slider object
ca_drop = california.drop(columns=['geometry']).copy()
#ca_source = ColumnDataSource(data=ca_drop)

# Create figure object.
p = figure(title = 'California by political party affiliation',
           plot_height = 300,
           plot_width = 250)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
```

```python
# Define color palettes
palette = brewer['RdBu'][10]

# Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = LinearColorMapper(palette = palette, low = 0, high = 100)

# Define custom tick labels for color bar.
tick_labels = {"0.0": "0%", "0.1": "10%", "0.2": "20%", "0.3": "30%",
"0.4": "40%", "0.5": "50%", "0.6": "60%", "0.7": "70%", "0.8": "80%", 
"0.9": "90%", "1.0": "100%"}

# Create color bar.
color_bar = ColorBar(color_mapper = color_mapper,
                     label_standoff = 10,
                     width = 500, height = 20,
                     border_line_color = None,
                     location = (0,0), 
                     orientation = "horizontal",
                     major_label_overrides = tick_labels)

states = p.patches("xs","ys", source = ca_source,
                   fill_color = {"field" :'perc_red',
                                 "transform" : color_mapper},
                   line_color = "black",
                   line_width = 0.25,
                   fill_alpha = 1)
```

```python
# Make a slider object to toggle the year shown
slider = Slider(title = "Year",
                start = 1984, end = 2016,
                step = 4, value = 2016)

# Creates custom filter that selects the rows of the month based on the value in the slider
custom_filter = CustomJSFilter(args = dict(slider = slider),
                               code = """
    var indices = [];
    for (var i = 0; i < source.get_length(); i++){
        if (source.data[year][i] == slider.value){
            indices.push(true);
        } else {
            indices.push(false);
        }
    }
    return indices;
""")

# Uses custom_filter to determine which set of sites are visible
view = CDSView(source = ca_source, filters = [custom_filter])

slider.js_on_change('value', CustomJS(args=dict(source=ca_source), code="""
   source.change.emit()
"""))
```

```python
p.add_tools(HoverTool(renderers = [states],
                      tooltips = [("County", '@NAMELSAD'),
                                    ('Year', "@year"), 
                                    ('% Republican Votes','@perc_red'),
                                ('% Democratic Votes','@perc_blue')]))

plot = column(p, slider)
#plot = p
```
